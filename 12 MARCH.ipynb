```
Q1.Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.
ANS-import scipy.stats as stats

n = 1 # Sample size
mean = 50
sd = 5
z = stats.norm.ppf(0.975) # 95% confidence level

lower_bound = mean - z * (sd / n**0.5)
upper_bound = mean + z * (sd / n**0.5)

print("95% confidence interval: [{:.2f}, {:.2f}]".format(lower_bound, upper_bound))


Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.
ANS-import numpy as np
import scipy.stats as stats

# Observed counts of M&M colors
observed = np.array([23, 17, 18, 11, 13, 18])

# Expected counts of M&M colors (20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown)
expected = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(observed)

# Calculate chi-square statistic and p-value
statistic, p_value = stats.chisquare(observed, expected)

# Print results
print("Chi-square statistic:", statistic)
print("p-value:", p_value)

if p_value < 0.05:
    print("Reject null hypothesis: The distribution of colors of M&Ms in the bag does not match the expected distribution.")
else:
    print("Fail to reject null hypothesis: The distribution of colors of M&Ms in the bag matches the expected distribution.")

Q3.
 Use Python to calculate the chi-square statistic and p-value for a contingency table with the following
data:
 

Interpret the results of the test.
ANS-import numpy as np
import scipy.stats as stats

# Contingency table
observed = np.array([[15, 20, 10], [30, 40, 20]])

# Calculate expected frequencies assuming independence
row_margins = np.sum(observed, axis=1)
col_margins = np.sum(observed, axis=0)
total = np.sum(row_margins)
expected = np.outer(row_margins, col_margins) / total

# Calculate chi-square statistic and p-value
statistic, p_value, _, _ = stats.chi2_contingency(observed)

# Print results
print("Chi-square statistic:", statistic)
print("p-value:", p_value)

if p_value < 0.05:
    print("Reject null hypothesis: The variables are not independent.")
else:
    print("Fail to reject null hypothesis: The variables are independent.")


Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.
ANS-import statsmodels.stats.proportion as proportion

# Sample size and number of successes
n = 500
successes = 60

# Calculate confidence interval
confint = proportion.proportion_confint(successes, n, alpha=0.05, method='wilson')

# Print results
print("95% confidence interval:", confint)


Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.
ANS-import scipy.stats as stats

# Sample statistics and sample size
mean = 75
std = 12
n = 50

# Degrees of freedom (n-1)
df = n - 1

# Calculate the standard error
se = std / (n ** 0.5)

# Calculate the confidence interval
confint = stats.t.interval(0.90, df, loc=mean, scale=se)

# Print the confidence interval
print("90% confidence interval:", confint)
Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.
ANS-import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# Degrees of freedom
df = 10

# Create x-axis values
x = np.linspace(0, 30, 1000)

# Create the chi-square distribution
chi2 = stats.chi2.pdf(x, df)

# Shade the area corresponding to a chi-square statistic of 15
plt.fill_between(x[x >= 15], chi2[x >= 15], color='red', alpha=0.5)

# Plot the chi-square distribution
plt.plot(x, chi2)

# Add axis labels and title
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')
plt.title('Chi-square distribution with 10 degrees of freedom')

# Show the plot
plt.show()

Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.
ANS-CI = (p-h, p+h)
where p is the sample proportion, h is the margin of error, and CI is the confidence interval.

The margin of error can be calculated as:
h = z*sqrt(p*(1-p)/n)
where z is the z-score corresponding to the desired confidence level, p is the sample proportion, and n is the sample size.

Using Python, we can calculate the confidence interval as follows
import math
import scipy.stats as stats

# Sample proportion
p = 520/1000

# Sample size
n = 1000

# Z-score for 99% confidence level
z = stats.norm.ppf(0.995)

# Margin of error
h = z*math.sqrt(p*(1-p)/n)

# Confidence interval
CI = (p-h, p+h)

print("99% Confidence interval:", CI)

Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.
ANS-To conduct a chi-square goodness of fit test in Python, we first need to define our null and alternative hypotheses.

Null hypothesis: The coin is fair and unbiased towards heads or tails.
Alternative hypothesis: The coin is biased towards tails.

We will use a significance level of 0.05, which means we will reject the null hypothesis if the p-value is less than 0.05.

To perform the test in Python, we can use the chisquare function from the scipy.stats module.
from scipy.stats import chisquare

# Observed frequencies
observed = [55, 45]

# Expected frequencies under the null hypothesis of a fair coin
expected = [50, 50]

# Calculate the chi-square statistic and p-value
statistic, p_value = chisquare(observed, expected)

# Print the results
print("Chi-square statistic:", statistic)
print("p-value:", p_value)

if p_value < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
Interpretation: The chi-square statistic is 1.0 and the p-value is 0.3173, which is greater than 0.05. Therefore, we fail to reject the null hypothesis and conclude that there is not enough evidence to suggest that the coin is biased towards tails.


Q9. A study was conducted to determine if there is an association between smoking status (smoker or non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below. Conduct a chi-square test for independence to determine if there is a significant association between smoking status and lung cancer diagnosis.Use a significance level of 0.05.
Group A
Outcome 1 20 15
Outcome 2 10 25
Outcome 3 15 20
Group B
Lung Cancer: Yes
Smoker 60 140
Non-smoker 30 170
ANS-import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# create contingency table
data = np.array([[20, 15], [10, 25], [15, 20], [60, 140], [30, 170]])
contingency_table = pd.DataFrame(data, index=['Outcome 1', 'Outcome 2', 'Outcome 3', 'Lung Cancer: Yes', 'Lung Cancer: No'], columns=['Smoker', 'Non-smoker'])
chi2, p_value, dof, expected = chi2_contingency(contingency_table)
 Q10.A study was conducted to determine if the proportion of people who prefer milk chocolate, dark chocolate, or white chocolate is different in the U.S. versus the U.K. A random sample of 500 people from the U.S. and a random sample of 500 people from the U.K. were surveyed. The results are shown in the contingency table below. Conduct a chi-square test for independence to determine if there is a significant association between chocolate preference and country of origin. Use a significance level of 0.01.
 
ANS-import numpy as np
from scipy.stats import chi2_contingency

# create contingency table
obs = np.array([[100, 200, 200],
                [150, 100, 250]])

# conduct chi-square test
chi2, p, dof, exp = chi2_contingency(obs)

# print results
print("Chi-square statistic:", chi2)
print("P-value:", p)

Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.
ANS-To conduct the hypothesis test, we can use a one-sample t-test with the null hypothesis that the population mean is equal to 70 and the alternative hypothesis that the population mean is not equal to 70. We will use a significance level of 0.05.

First, we calculate the t-statistic:

t = (72 - 70) / (10 / sqrt(30))
t = 1.897

Next, we calculate the degrees of freedom:

df = n - 1 = 29

Using a t-distribution table or a t-distribution calculator, we find that the p-value for a two-tailed test with 29 degrees of freedom and a t-statistic of 1.897 is approximately 0.067.

Since the p-value (0.067) is greater than the significance level (0.05), we fail to reject the null hypothesis. Therefore, we do not have sufficient evidence to conclude that the population mean is significantly different from 70 at the 0.05 significance level.
```
