````
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
ANS-A point estimate is a single value estimate of a parameter. For instance, a sample mean is a point estimate of a population mean. An interval estimate gives you a range of values where the parameter is expected to lie. A confidence interval is the most common type of interval estimate.
A point estimate of a parameter θ is a value (based on a sample) that is a sensible guess for θ. A point estimate is obtained by a formula (“estimator”) which takes the sample data and produces an point estimate. Such formulas are called point estimators of θ.


Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.
ANS-The statistics module in python provides functions called stdev() and pstdev() to calculate the standard deviation of a sample dataset. The stdev() function in python only calculates the sample standard deviation whereas the pstdev() function calculates the population standard deviation.
Python Scipy stats module can be used to create a normal distribution with meand and standard deviation parameters using method norm. Standard normal distribution is normal distribution with mean as 0 and standard deviation as 1.
Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.
ANS-Hypothesis testing is used to assess the plausibility of a hypothesis by using sample data. The test provides evidence concerning the plausibility of the hypothesis, given the data. Statistical analysts test a hypothesis by measuring and examining a random sample of the population being analyzed.
Hypothesis testing is the process used to evaluate the strength of evidence from the sample and provides a framework for making determinations related to the population, ie, it provides a method for understanding how reliably one can extrapolate observed findings in a sample under study to the larger population from 
After developing your initial research hypothesis (the prediction that you want to investigate), it is important to restate it as a null (Ho) and alternate (Ha) hypothesis so that you can test it mathematically. The alternate hypothesis is usually your initial hypothesis that predicts a relationship between variables
Q4. Create a hypothesis that states whether the average weight of male college students is greater than the average weight of female college students.
ANS-A z-test is used to test a Null Hypothesis if the population variance is known, or if the sample size is larger than 30, for an unknown population variance. A t-test is used when the sample size is less than 30 and the population variance is unknown
Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.
ANS-import numpy as np
from scipy.stats import ttest_ind

# define the two samples
sample1 = np.array([78, 72, 85, 66, 77, 80, 82, 68, 74, 81])
sample2 = np.array([62, 73, 71, 69, 64, 75, 70, 68, 72, 67])

# calculate the mean and standard deviation of each sample
mean1 = np.mean(sample1)
std1 = np.std(sample1, ddof=1)
mean2 = np.mean(sample2)
std2 = np.std(sample2, ddof=1)

# calculate the t-statistic and p-value
t, p = ttest_ind(sample1, sample2, equal_var=False)

# print the results
print("Sample 1 mean: {:.2f}".format(mean1))
print("Sample 1 standard deviation: {:.2f}".format(std1))
print("Sample 2 mean: {:.2f}".format(mean2))
print("Sample 2 standard deviation: {:.2f}".format(std2))
print("t-statistic: {:.2f}".format(t))
print("p-value: {:.4f}".format(p))
Q6: What is a null and alternative hypothesis? Give some examples.
ANS-A null hypothesis is a statement about a population parameter that we wish to test for statistical significance. It is often denoted as H0 and represents the status quo or default position, assuming that there is no significant difference or relationship between the variables being tested.

An alternative hypothesis, denoted as Ha, is a statement that contradicts the null hypothesis, and represents the possibility that there is a significant difference or relationship between the variables being tested.

Here are some examples of null and alternative hypotheses:

Example for a one-sample hypothesis test:
Null hypothesis: The mean IQ score of a population is 100.
Alternative hypothesis: The mean IQ score of a population is not 100.

Example for a two-sample hypothesis test:
Null hypothesis: The mean height of men is the same as the mean height of women.
Alternative hypothesis: The mean height of men is different from the mean height of women.

Example for a correlation hypothesis test:
Null hypothesis: There is no significant correlation between age and income.
Alternative hypothesis: There is a significant correlation between age and income.

In each of these examples, the null hypothesis represents the absence of a significant effect or relationship, while the alternative hypothesis represents the possibility of a significant effect or relationship. The hypothesis test is designed to determine which of these two hypotheses is supported by the evidence in the data.
Q7: Write down the steps involved in hypothesis testing.
ANS-Here are the general steps involved in hypothesis testing:

1. State the null and alternative hypotheses: Clearly define the null and alternative hypotheses based on the research question and prior knowledge.

2. Select the significance level: Choose the level of significance (alpha), typically set at 0.05, which represents the probability of rejecting the null hypothesis when it is true.

3. Choose the appropriate test statistic: Select the appropriate test statistic based on the nature of the hypothesis test, sample size, and type of data.

4. Collect and analyze the data: Collect the sample data and perform the necessary calculations to obtain the test statistic and the p-value.

5. Make a decision: Compare the p-value to the significance level, and determine whether to reject or fail to reject the null hypothesis.

6. Draw conclusions: Interpret the results of the hypothesis test, and draw conclusions about the population parameter of interest.

7. Communicate the results: Report the results of the hypothesis test, including the test statistic, p-value, and conclusion, in a clear and concise manner.

It's important to note that the specific steps involved in hypothesis testing may vary depending on the type of hypothesis test being conducted and the statistical software or programming language being used. However, these general steps provide a framework for conducting hypothesis tests and making statistical inferences based on sample data.
Q8. Define p-value and explain its significance in hypothesis testing.
ANS-The p-value is a measure of the strength of evidence against the null hypothesis in a hypothesis test. It represents the probability of obtaining a test statistic at least as extreme as the one observed, assuming the null hypothesis is true.

In other words, the p-value tells us how likely it is to observe the data we have if the null hypothesis is true. A small p-value indicates that it is unlikely to observe such data if the null hypothesis is true, and therefore provides evidence against the null hypothesis. Conversely, a large p-value suggests that the data are consistent with the null hypothesis, and there is insufficient evidence to reject it.

The significance level (alpha) is typically set at 0.05, which means that if the p-value is less than 0.05, we can reject the null hypothesis and conclude that there is a statistically significant difference or relationship between the variables being tested. If the p-value is greater than 0.05, we fail to reject the null hypothesis and conclude that there is not enough evidence to support the alternative hypothesis.

It's important to note that the p-value does not tell us the size or practical importance of the effect or relationship being tested, but rather indicates whether the effect or relationship is statistically significant or not. Additionally, the p-value is affected by the sample size, significance level, and assumptions of the statistical test being used, so it should be interpreted carefully in the context of the specific hypothesis test being conducted.

Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom parameter set to 10.
ANS-Here's an example Python code to generate a Student's t-distribution plot using matplotlib library with degrees of freedom parameter set to 10:

```
import numpy as np
import matplotlib.pyplot as plt

# Generate data
df = 10
x = np.linspace(-4, 4, 100)
y = stats.t.pdf(x, df)

# Create plot
fig, ax = plt.subplots()
ax.plot(x, y, label='t-distribution')

# Set plot properties
ax.set_xlabel('x')
ax.set_ylabel('pdf')
ax.set_title('Student\'s t-distribution with df = {}'.format(df))
ax.legend()

plt.show()
```

This code first imports numpy and matplotlib.pyplot libraries, then generates data for the t-distribution using the stats.t.pdf() function from scipy.stats library. The degrees of freedom parameter is set to 10 in this case. 

Next, the code creates a plot using matplotlib's subplots() function, then plots the t-distribution using the plot() function on the generated data.

Finally, the plot is customized with labels and titles for the x-axis, y-axis, and plot title using the set_xlabel(), set_ylabel(), and set_title() functions respectively, and the plot is displayed using the show() function.
Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.
ANS-import numpy as np
from scipy.stats import ttest_ind

# Generate two random samples with equal size
sample1 = np.random.normal(10, 2, 50)
sample2 = np.random.normal(12, 2, 50)

# Calculate the t-statistic and p-value for the two-sample t-test
t_stat, p_value = ttest_ind(sample1, sample2, equal_var=True)

# Print the results
print('t-statistic: ', t_stat)
print('p-value: ', p_value)
In this code, we first import numpy and scipy.stats libraries. We then generate two random samples of equal size using numpy's random.normal() function, with mean 10 and 12, standard deviation 2, and sample size 50.

Next, we use the ttest_ind() function from scipy.stats library to calculate the t-statistic and p-value for the two-sample t-test. The equal_var parameter is set to True, assuming equal variances in both samples.

Finally, the code prints the calculated t-statistic and p-value to the console using the print() function. Depending on the calculated values, we can draw conclusions about whether or not to reject the null hypothesis that the population means are equal.
Q11: What is Student’s t distribution? When to use the t-Distribution.
ANS-Student's t-distribution is a probability distribution that is used in hypothesis testing when the sample size is small or when the population variance is unknown. It is similar to the normal distribution but has heavier tails and more probability in the tails, which makes it more appropriate for small sample sizes.

The t-distribution is used when the population standard deviation is unknown and must be estimated from the sample standard deviation. It is also used when the sample size is small (typically less than 30) and the population distribution is not normal, but the sample is approximately normal.

The t-distribution is commonly used in hypothesis testing for testing the difference between two means, testing whether a sample mean is significantly different from a population mean, or testing the significance of a correlation coefficient. The t-distribution is also used in confidence interval estimation for a population mean when the sample size is small or the population variance is unknown.

In summary, the t-distribution is a statistical tool used in situations where the population variance is unknown and the sample size is small. It provides a way to estimate the population mean and assess the significance of the difference between sample means.
Q12: What is t-statistic? State the formula for t-statistic.
ANS-The t-statistic is a measure of how different the sample mean is from the hypothesized population mean, in units of the standard error of the sample mean. It is used to test hypotheses about the population mean when the population standard deviation is unknown and must be estimated from the sample.

The formula for the t-statistic is:

t = (x̄ - μ) / (s / sqrt(n))

Where:
- x̄ is the sample mean
- μ is the hypothesized population mean
- s is the sample standard deviation
- n is the sample size

The t-statistic is calculated by subtracting the hypothesized population mean from the sample mean, then dividing the result by the standard error of the sample mean. The standard error of the sample mean is the standard deviation of the sampling distribution of the mean, which is calculated as the sample standard deviation divided by the square root of the sample size.

The resulting t-statistic measures how many standard errors the sample mean is from the hypothesized population mean. If the t-statistic is large, it suggests that the sample mean is significantly different from the hypothesized population mean, and we may reject the null hypothesis that the sample mean is equal to the population mean. If the t-statistic is small, it suggests that the sample mean is not significantly different from the hypothesized population mean, and we may fail to reject the null hypothesis.

Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50. Estimate the population mean revenue with a 95% confidence interval.
ANS-To estimate the population mean revenue with a 95% confidence interval, we can use the t-distribution since the population standard deviation is unknown and the sample size is less than 30. 

The formula for the confidence interval for a population mean is:

Confidence Interval = Sample Mean ± Margin of Error

where the margin of error is:

Margin of Error = t * (Standard Deviation / sqrt(n))

where t is the critical value from the t-distribution for the desired confidence level, Standard Deviation is the sample standard deviation, and n is the sample size.

For a 95% confidence level and 49 degrees of freedom (50-1), the critical value from the t-distribution can be found using a t-table or Python's scipy.stats library. Using Python, we can calculate the confidence interval as:

```
import scipy.stats as stats

sample_mean = 500
sample_std = 50
n = 50
alpha = 0.05  # 1 - confidence level

t = stats.t.ppf(1-alpha/2, n-1)  # t-value for 95% confidence level with 49 degrees of freedom

margin_of_error = t * (sample_std / np.sqrt(n))

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("The 95% confidence interval for the population mean revenue is: $", confidence_interval)

Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.
ANS-To test the hypothesis that the new drug will decrease blood pressure by 10 mmHg with a significance level of 0.05, we can use a one-sample t-test. The null hypothesis is that the mean decrease in blood pressure is equal to 10 mmHg, and the alternative hypothesis is that the mean decrease in blood pressure is less than 10 mmHg.

The test statistic for the one-sample t-test is calculated as:

t = (sample_mean - hypothesized_mean) / (sample_std / sqrt(sample_size))

where sample_mean is the sample mean, hypothesized_mean is the hypothesized mean (10 mmHg in this case), sample_std is the sample standard deviation, and sample_size is the sample size.

The critical t-value for a one-tailed t-test with 99 degrees of freedom (100-1) at a significance level of 0.05 can be found using a t-table or Python's scipy.stats library. Using Python, we can calculate the test statistic and critical t-value as:

```
import scipy.stats as stats
import numpy as np

sample_mean = 8
hypothesized_mean = 10
sample_std = 3
sample_size = 100
alpha = 0.05  # significance level

t_statistic = (sample_mean - hypothesized_mean) / (sample_std / np.sqrt(sample_size))

t_critical = stats.t.ppf(alpha, sample_size-1)

print("t-statistic: ", t_statistic)
print("t-critical: ", t_critical)
```

Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5 pounds with a significance level of 0.01.
ANS-To test the hypothesis that the true mean weight of the products is less than 5 pounds with a significance level of 0.01, we can use a one-sample t-test. The null hypothesis is that the true mean weight is equal to 5 pounds, and the alternative hypothesis is that the true mean weight is less than 5 pounds.

The test statistic for the one-sample t-test is calculated as:

t = (sample_mean - hypothesized_mean) / (sample_std / sqrt(sample_size))

where sample_mean is the sample mean weight, hypothesized_mean is the hypothesized mean (5 pounds in this case), sample_std is the sample standard deviation, and sample_size is the sample size.

The critical t-value for a one-tailed t-test with 24 degrees of freedom (25-1) at a significance level of 0.01 can be found using a t-table or Python's scipy.stats library. Using Python, we can calculate the test statistic and critical t-value as:

```
import scipy.stats as stats
import numpy as np

sample_mean = 4.8
hypothesized_mean = 5
sample_std = 0.5
sample_size = 25
alpha = 0.01  # significance level

t_statistic = (sample_mean - hypothesized_mean) / (sample_std / np.sqrt(sample_size))

t_critical = stats.t.ppf(alpha, sample_size-1)

print("t-statistic: ", t_statistic)
print("t-critical: ", t_critical)
```


Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two groups are equal with a significance level of 0.01.
ANS-To test the hypothesis that the population means for the two groups are equal with a significance level of 0.01, we can use a two-sample t-test assuming unequal variances. The null hypothesis is that the population means are equal, and the alternative hypothesis is that they are not equal.

The test statistic for the two-sample t-test assuming unequal variances is calculated as:

t = (sample_mean1 - sample_mean2) / sqrt((s1^2/n1) + (s2^2/n2))

where sample_mean1 and sample_mean2 are the sample means, s1 and s2 are the sample standard deviations, n1 and n2 are the sample sizes.

The degrees of freedom for the t-test can be calculated using the following formula:

df = ((s1^2/n1) + (s2^2/n2))^2 / (((s1^2/n1)^2)/(n1-1) + ((s2^2/n2)^2)/(n2-1))

Using Python, we can calculate the test statistic and the degrees of freedom as:

```
import numpy as np
import scipy.stats as stats

n1 = 30
n2 = 40
mean1 = 80
mean2 = 75
std1 = 10
std2 = 8
alpha = 0.01

s_p = np.sqrt(((n1-1)*(std1**2) + (n2-1)*(std2**2)) / (n1+n2-2))
t_statistic = (mean1 - mean2) / np.sqrt((std1**2/n1) + (std2**2/n2))
df = ((std1**2/n1) + (std2**2/n2))**2 / (((std1**2/n1)**2)/(n1-1) + ((std2**2/n2)**2)/(n2-1))

p_value = 2 * stats.t.cdf(-abs(t_statistic), df)

print("t-statistic: ", t_statistic)
print("degrees of freedom: ", df)
print("p-value: ", p_value)

if p_value < alpha:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```


Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard deviation of 1.5. Estimate the population mean with a 99% confidence interval.
ANS-Sample size (n) = 50
Sample mean (x̄) = 4
Sample standard deviation (s) = 1.5
We need to calculate the 99% confidence interval for the population mean, μ.
The formula for calculating the confidence interval is:
CI = x̄ ± z* (s/√n)
Where z* is the z-score for the desired confidence level. For a 99% confidence level, the z-score is 2.576.
Plugging in the values, we get:
CI = 4 ± 2.576 * (1.5/√50)
CI = 4 ± 0.640
So the 99% confidence interval for the population mean is (3.36, 4.64).
Therefore, we can say with 99% confidence that the true population mean of ads watched by viewers during a TV program lies between 3.36 and 4.64.
````
