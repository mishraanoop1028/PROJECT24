```
Q1.ARTIFICAL VS DEEP VS MACHINE
ANS=Artificial Intelligence (AI) is a broad field of computer science that deals with the creation of intelligent machines capable of performing tasks that typically require human intelligence. One of the subfields of AI is Machine Learning (ML), which is the process of training machines to learn patterns and make predictions from data without being explicitly programmed. Deep Learning (DL) is a subset of ML that uses neural networks with many layers to learn complex representations of data. 

Machine Learning (ML) is the process of training machines to learn patterns and make predictions from data without being explicitly programmed. It involves the use of statistical techniques to enable machines to learn from data, identify patterns, and make predictions or decisions. For example, an ML model could be trained to predict whether a customer is likely to purchase a product based on their purchase history and demographic data.

Example: An e-commerce website uses ML to predict which products a customer is most likely to purchase based on their browsing and purchase history.

Deep Learning (DL) is a subset of ML that uses neural networks with many layers to learn complex representations of data. DL is particularly well-suited for tasks that require the identification of complex patterns in large datasets. For example, DL has been used in image and speech recognition applications.

Example: A DL model is trained to recognize handwritten digits in images.

Q2- SUPERVISED 
ANS-Supervised Learning is a type of machine learning where the algorithm learns to make predictions based on labeled data. In supervised learning, the algorithm is provided with a dataset that includes both input features and output labels, and the algorithm learns to map the input features to the output labels. The goal of supervised learning is to build a model that can accurately predict the output label for new, unseen input data.

Some examples of supervised learning include:

1. Image classification: Given a dataset of images labeled with their corresponding object classes, an algorithm can be trained to recognize new images and correctly classify them into their respective categories.

2. Sentiment analysis: Given a dataset of text labeled with positive or negative sentiment, an algorithm can be trained to classify new text as having a positive or negative sentiment.

3. Fraud detection: Given a dataset of transactions labeled as fraudulent or not fraudulent, an algorithm can be trained to detect new fraudulent transactions.

4. Spam filtering: Given a dataset of emails labeled as spam or not spam, an algorithm can be trained to filter out new spam emails.
Q3.WHAT IS UNSUPERVISED LEARNING AND ITS EXAPLE ?
ANS-Unsupervised learning is a type of machine learning where the algorithm learns to identify patterns and relationships in data without being explicitly told what to look for. Unlike supervised learning, unsupervised learning doesn't have labeled output data to guide the learning process. Instead, the algorithm identifies patterns and structures in the data on its own, without any prior knowledge of the output labels.

Some examples of unsupervised learning include:

1. Clustering: Given a dataset of data points, an algorithm can be used to group together similar data points into clusters, based on their similarities in terms of features.

2. Anomaly detection: Given a dataset of data points, an algorithm can be used to identify unusual or anomalous data points that don't fit into any of the expected patterns or clusters.

3. Dimensionality reduction: Given a dataset of high-dimensional data, an algorithm can be used to reduce the dimensionality of the data while retaining the most important features and patterns.

4. Association rule learning: Given a dataset of transaction data, an algorithm can be used to identify patterns and relationships between different items or products, such as which products are frequently purchased together.
Q.4 ML,DS,DL,AL 
ANS-AI (Artificial Intelligence), ML (Machine Learning), DL (Deep Learning), and DS (Data Science) are related fields, but they differ in their scope, approach, and application. Here are the key differences between these fields:

1. AI: AI is a broad field that focuses on creating intelligent machines that can perform tasks that typically require human intelligence. It encompasses various techniques and approaches, including rule-based systems, expert systems, decision trees, and machine learning.

2. ML: ML is a subset of AI that focuses on designing algorithms and models that can learn from data and improve their performance without being explicitly programmed. It includes various techniques, such as regression, decision trees, neural networks, and support vector machines.

3. DL: DL is a subset of ML that uses deep neural networks to learn from large amounts of data. It requires a significant amount of data and computing power to train deep neural networks.

4. DS: DS is an interdisciplinary field that focuses on extracting insights and knowledge from data using various techniques, such as data mining, statistics, machine learning, and visualization. It involves collecting, cleaning, processing, analyzing, and interpreting data to gain insights and inform decision-making.

In summary, AI is a broad field that includes ML and DL, while DS is a related field that focuses on extracting insights from data using various techniques, including machine learning.

Q5.WHAT IS THE differences between supervised, unsupervised, and semi-supervised learning ?
ANS NO -5 
The main differences between supervised, unsupervised, and semi-supervised learning are:

1. Supervised Learning: In supervised learning, the model is trained on labeled data, which means that the input data is paired with the correct output. The goal of supervised learning is to learn a mapping between the input and output variables, so that the model can predict the output for new input data.

2. Unsupervised Learning: In unsupervised learning, the model is trained on unlabeled data, which means that the input data is not paired with the correct output. The goal of unsupervised learning is to discover patterns, structures, and relationships in the data, without any prior knowledge of what the output should be.

3. Semi-supervised Learning: Semi-supervised learning is a combination of supervised and unsupervised learning. In semi-supervised learning, the model is trained on both labeled and unlabeled data, which allows it to leverage the benefits of both approaches. Semi-supervised learning is useful when labeled data is scarce or expensive to obtain.

The main differences between these three types of learning can be summarized as follows:

- Supervised learning requires labeled data, while unsupervised learning does not require labeled data.
- Supervised learning is used for prediction tasks, while unsupervised learning is used for exploration tasks.
- Semi-supervised learning is used when there is a small amount of labeled data and a large amount of unlabeled data, which is often the case in real-world applications.
- Supervised learning is easier to evaluate and measure performance, while unsupervised learning is more challenging to evaluate and requires domain knowledge and expertise.
- Semi-supervised learning can often achieve better performance than supervised learning when there is a limited amount of labeled data available.
Q6.WHAT IS TRAIN TEST AND VALIDATION SPLIT AND ITS EXAPLE ?
ANS-In machine learning, it is common practice to split the available data into three subsets: training set, validation set, and test set. The purpose of this split is to evaluate and select the best model for a given task.

- Training Set: The training set is the subset of data used to fit the parameters of the machine learning model. In other words, the model learns from the data in the training set to make predictions on new, unseen data. 

- Validation Set: The validation set is the subset of data used to tune the hyperparameters of the machine learning model. Hyperparameters are the parameters that are not learned from the data, but instead, they control the learning process of the model. By tuning the hyperparameters on the validation set, we can find the optimal configuration of the model that gives the best performance on the test set.

- Test Set: The test set is the subset of data used to evaluate the performance of the machine learning model. It is a completely new, unseen data that the model has not seen during training or validation. The test set gives an unbiased estimate of the model's performance on new data.

The reason for splitting the data into three subsets is to prevent overfitting of the model. Overfitting occurs when a model learns the noise in the training data, resulting in poor performance on new, unseen data. By using a separate validation set to tune the hyperparameters, we can avoid overfitting the model to the training data. Finally, we use the test set to evaluate the model's performance on completely new data, which gives an unbiased estimate of the model's ability to generalize to new situations.
Q7.HOW UNSUPERVISED LEARNING USED ?

ANS-Unsupervised learning can be used in anomaly detection by identifying patterns in the data and flagging any deviations from those patterns as anomalies. One common technique is to use clustering algorithms to group similar data points together and identify any points that do not fit into any cluster as anomalies. Another approach is to use density-based anomaly detection, which identifies anomalies as data points that have a significantly lower density compared to the surrounding data points. Overall, unsupervised learning provides a useful tool for detecting anomalies in data without the need for labeled examples.
Q8.LIST DOWN SUPERVISED AND UNSUPERVISED LEARNING AND ITS ALGORITHM?
ANS-Some commonly used supervised learning algorithms are:
- Linear Regression
- Logistic Regression
- Decision Trees
- Random Forest
- Support Vector Machines (SVM)
- Neural Networks

Some commonly used unsupervised learning algorithms are:
- K-Means Clustering
- DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
- PCA (Principal Component Analysis)
- Anomaly Detection
- Association Rule Mining
- Gaussian Mixture Models (GMM)
```
